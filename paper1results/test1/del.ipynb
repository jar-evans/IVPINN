```python
import numpy as np
import math

import tensorflow as tf
import numpy as np

def u_ex(x, y):
    return np.tanh(2*(x**3 - y**4))

def f(x, y):
    tmp = np.tanh(2*(x**3 - y**4))
    A = -12*x*(1-6*tmp*x**3)*(np.sech(tmp))**2
    return A - 8*y*y*(16*tmp*y**4 + 3)*(np.sech(tmp))**2
```

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np

## Add surface plot
N = 50
x=np.linspace(0,1,N,dtype=np.float32)
XX,YY = np.meshgrid(x,x)

ZZ = u_ex(XX, YY)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
surf = ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap=cm.coolwarm,
    linewidth=0, antialiased=False)
# ax.set_zlim(-1.01, 0)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)
plt.title('NN solution')
```
