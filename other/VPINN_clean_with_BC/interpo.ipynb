```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 1, 4, 3, 5])

# Define points to interpolate
x_interp = 2.5

# Perform linear interpolation
y_interp = np.interp(x_interp, x, y)

# Plot the original data points
plt.scatter(x, y, label='Original Data')

# Plot the interpolated point
plt.scatter(x_interp, y_interp, color='red', label='Interpolated Point')

# Connect the original points with lines
plt.plot(x, y, '--', label='Interpolation Lines')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Show the plot
plt.show()
```
